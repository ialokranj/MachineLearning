```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
df
```

```python
target = df.target
df = df.drop('target', axis = 'columns')
```

```python
df
```

```python
target
```

DataSet for PCA

```python
X = df
y = target
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train_scaled, x_test_scaled, y_train_scaled, y_test_scaled = train_test_split(X_scaled, y, train_size = 0.2)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

# Building a Model using Linear Regression without PCA

DataSet for Non-PCA

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df, target, test_size = 0.2)
```

```python
x_train
```

```python
x_test
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

# Building a Model using Linear Regression with PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=4)
X_pca = pca.fit_transform(X)
```

```python
x_train_pca, x_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr_pca = LinearRegression()
lr_pca.fit(x_train_pca, y_train_pca)
lr_pca.score(x_test_pca, y_test_pca)
```

# Building a Model using Logistic Regression without PCA

```python
from sklearn.linear_model import LogisticRegression
logi = LogisticRegression()
logi.fit(x_train, y_train)
```

```python
logi.score(x_test, y_test)
```

# Building a Model using Logistic Regression with PCA

```python
logi_pca = LogisticRegression()
logi_pca.fit(x_train_pca, y_train_pca)
logi_pca.score(x_test_pca, y_test_pca)
```

# Building a Model using Random Forest Classifier without PCA

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 40)
rfc.fit(x_train, y_train)
```

```python
rfc.score(x_test, y_test)
```

# Building a Model using Random Forest Classifier with PCA

```python
rfc_pca = RandomForestClassifier(n_estimators = 40)
rfc_pca.fit(x_train_pca, y_train_pca)
rfc_pca.score(x_test_pca, y_test_pca)
```
